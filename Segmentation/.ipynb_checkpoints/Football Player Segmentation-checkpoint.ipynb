```python
import fiftyone as fo
import fiftyone.utils.huggingface as fouh

# Load the dataset

dataset = fouh.load_from_hub("voxel51/Football-Player-Segmentation")
```

```python
session = fo.launch_app(dataset)
```

```python
import fiftyone.zoo as foz

model = foz.load_zoo_model("segment-anything-vitb-torch")

# Segment inside boxes
dataset.apply_model(
    model,
    label_field="Guided-Segment-Anything",
    prompt_field="detections",  # can contain Detections or Keypoints
)
```

```python
`session.show()
```

```python
results = dataset.evaluate_detections(
    "Guided-Segment-Anything", gt_field="segmentations", eval_key="eval", use_masks = True
)
```

```python
# Convert to evaluation patches
eval_patches = dataset.to_evaluation_patches("eval")
print(eval_patches)
```

```python
print(eval_patches.count_values("type"))
```

```python
# View patches in the App
session.view = eval_patches
```
