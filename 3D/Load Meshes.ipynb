```python
import fiftyone as fo

mesh_dir = "/home/dan/Documents/datasets/3d-test/"

dataset = fo.Dataset("3D-Mesh-Example")
```

```python
ply_path = mesh_dir + "house.ply"
    
# We load in with PlyMesh and pass our files in
ply = fo.PlyMesh(name="ply-example", ply_path=ply_path)

# We define our scene and save it so it can be loaded into FiftyOne
scene = fo.Scene(camera=fo.PerspectiveCamera(up="Z"))
scene.add(ply)
scene_path = mesh_dir + "ply-example.fo3d"
scene.write(scene_path)

# Finally we add the scene as a sample
mesh_sample = fo.Sample(scene_path)
dataset.add_sample(mesh_sample)

session = fo.launch_app(dataset)
```
